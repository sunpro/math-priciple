
## 逻辑回归
逻辑回归的目的：分类  
因此就要找一个函数，能够将输入映射为一个有较大差别的输出~类似于这样：

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(-10,0,0.1)
x2 = np.arange(0,10,0.1)
y1 = np.zeros((100,));
y2 = np.ones((100,));
#print(x1)
#print(y1)
plt.plot(x1,y1,'r')
plt.plot(x2,y2,'b')
plt.axvline(0,color='r',linestyle='--')
plt.axis([-10, 10, -0.2, 1.2]) 
plt.show()
```

逻辑回归的公式：
$$h_\theta(x)=\frac{1}{1+\exp(x)}$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-6,6,0.1)
y = 1/(1+np.exp(-x))
#print(y)
plt.plot(x,y)
plt.axhline(0.5,color='r',linestyle='--')
plt.axvline(0,color='r',linestyle='--')
plt.show()
```
